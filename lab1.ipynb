```python
def log_profitability(df):
    data_profit = df
    for column in range(0,df.shape[1]):
        for index in range(0,(df.shape[0])-1):
            data_profit.iloc[index][column] = np.log(df.iloc[index+1][column]/df.iloc[index][column])
    data_profit.drop(data_profit.tail(1).index,inplace=True)
    #data_profit = data_profit.drop(data_profit.iloc[-1:]) 
    #нужно дропнуть последнюю сроку
    return data_profit
```

```python
log_profit1 = log_profitability(adj_close)
```
