### Загружаем данные

```python
from pandas_datareader import data
```

```python
data_source = 'yahoo'
start_date = '2010-01-02'
end_date = '2014-12-31'
```

```python
panel_data = data.DataReader(namesNew, data_source, start_date, end_date)
```

```python
panel_data
```

```python
a = copy.copy(panel_data)
```

```python
adj_close = a.loc['Adj Close']
```

```python
adj_close
```

```python
adj_close.describe()
```

```python
adj_close = adj_close.drop(pd.Timestamp('2010-12-24'))
adj_close = adj_close.drop(pd.Timestamp('2012-12-24'))
adj_close = adj_close.dropna(axis = 0)
adj_close.shape
```

```python
adj_close.to_csv(path_or_buf='adj_close.csv', sep=';')
```
